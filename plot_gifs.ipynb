### Create GIF

```python
import os
import re
# import imageio
import imageio.v2 as imageio

# Define the directory and regex pattern
directory = r'D:\Work\Box Sync\Quantify Infrastructure\Figures\below_10000'
pattern = re.compile(r'^temporalPerCapByState__below_\d{4}\.png$')  # Example pattern: file_001.txt, file_002.txt, etc.

# List and filter files
png_files = [f for f in os.listdir(directory) if pattern.match(f)]
# List of image file paths
images = []

# Process the matching files
for file_name in png_files:
    filename = os.path.join(directory, file_name)
    images.append(imageio.imread(filename))

# Save as GIF
imageio.mimsave(r'D:\Work\Box Sync\Quantify Infrastructure\Figures\below_10000\below_10000.gif', images, fps=2)
png_files
```
